```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data.info()
#change the data type of order date from object to datetime
```

```python
year2015 = data[data['Order Date'].dt.year == 2015]
year2016 = data[data['Order Date'].dt.year == 2016]
year2017 = data[data['Order Date'].dt.year == 2017]
year2018 = data[data['Order Date'].dt.year == 2018]

x = np.arange(1, 13)

sales2015 = year2015.groupby(year2015['Order Date'].dt.month)['Sales'].sum()
sales2016 = year2016.groupby(year2016['Order Date'].dt.month)['Sales'].sum()
sales2017 = year2017.groupby(year2017['Order Date'].dt.month)['Sales'].sum()
sales2018 = year2018.groupby(year2018['Order Date'].dt.month)['Sales'].sum()

plt.figure(figsize=(10, 6))
plt.bar(x-0.3, sales2015, width=0.2, label='2015')
plt.bar(x-0.1, sales2016, width=0.2, label='2016')
plt.bar(x+0.1, sales2017, width=0.2, label='2017')
plt.bar(x+0.3, sales2018, width=0.2, label='2018')
plt.legend(title='Year')
plt.title('Multiple bar chart of monthly gross sales for different years')
plt.xlabel('Month')
plt.ylabel('Gross Sales', rotation=90)
plt.xticks(x, data['Order Date'].dt.month.value_counts().sort_index(ascending=True).index)

plt.show()
# y axis is sum of sales
```

```python
year2015 = data[data['Order Date'].dt.year == 2015]
year2016 = data[data['Order Date'].dt.year == 2016]
year2017 = data[data['Order Date'].dt.year == 2017]
year2018 = data[data['Order Date'].dt.year == 2018]

x = np.arange(1, 13)

profit2015 = year2015.groupby(year2015['Order Date'].dt.month)['Profit'].sum()
profit2016 = year2016.groupby(year2016['Order Date'].dt.month)['Profit'].sum()
profit2017 = year2017.groupby(year2017['Order Date'].dt.month)['Profit'].sum()
profit2018 = year2018.groupby(year2018['Order Date'].dt.month)['Profit'].sum()

plt.figure(figsize=(10, 6))

plt.bar(x-0.3, profit2015, width=0.2, label='2015')
plt.bar(x-0.1, profit2016, width=0.2, label='2016')
plt.bar(x+0.1, profit2017, width=0.2, label='2017')
plt.bar(x+0.3, profit2018, width=0.2, label='2018')

plt.legend(title='Year')
plt.title('Multiple bar chart of monthly profit for different years')
plt.xlabel('Month')
plt.ylabel('Profit', rotation=90)
plt.xticks(x, data['Order Date'].dt.month.value_counts().sort_index(ascending=True).index)

plt.show()
```

```python
year2015 = data[data['Order Date'].dt.year == 2015]
year2016 = data[data['Order Date'].dt.year == 2016]
year2017 = data[data['Order Date'].dt.year == 2017]
year2018 = data[data['Order Date'].dt.year == 2018]

x = np.arange(1, 13)

profit2015 = year2015.groupby(year2015['Order Date'].dt.month)['Profit'].sum()/year2015['Sales'].sum() * 100
profit2016 = year2016.groupby(year2016['Order Date'].dt.month)['Profit'].sum()/year2016['Sales'].sum() * 100
profit2017 = year2017.groupby(year2017['Order Date'].dt.month)['Profit'].sum()/year2017['Sales'].sum() * 100
profit2018 = year2018.groupby(year2018['Order Date'].dt.month)['Profit'].sum()/year2018['Sales'].sum() * 100

plt.figure(figsize=(10, 6))

plt.bar(x-0.3, profit2015, width=0.2, label='2015')
plt.bar(x-0.1, profit2016, width=0.2, label='2016')
plt.bar(x+0.1, profit2017, width=0.2, label='2017')
plt.bar(x+0.3, profit2018, width=0.2, label='2018')

plt.legend(title='Year')
plt.title('Multiple bar chart of monthly profit margin (%) for different years')
plt.xlabel('Month')
plt.ylabel('Profit Margin(%)', rotation=90)
plt.xticks(x, data['Order Date'].dt.month.value_counts().sort_index(ascending=True).index)

plt.show()

print(profit2015.sum() + profit2016.sum() + profit2017.sum() + profit2018.sum())
print(profit2015.sum(), profit2016.sum(), profit2017.sum(), profit2018.sum())
```
